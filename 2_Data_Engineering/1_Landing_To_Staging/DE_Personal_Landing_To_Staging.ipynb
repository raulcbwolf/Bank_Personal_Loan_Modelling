# Librerías

```python
import pandas as pd
```

# Rutas

```python
PERSONAL_INPUT_PATH = '../../1_Datasets/1_Landing/Personal_Data.csv'
```

# Lectura

```python
personal_df = pd.read_csv(PERSONAL_INPUT_PATH)
```

# Preprocesado

```python
personal_df = personal_df.rename(columns = {
    'ZIP Code': 'ZIPCode',
    'First Name': 'FirstName',
    'Last Name': 'LastName',
})

personal_df
```

```python
cond = (
    (personal_df['Age'] >= 0) &
    (~personal_df['Age'].isnull()) &
    (personal_df['Family'] >= 0) &
    (~personal_df['Family'].isnull()) &
    (~personal_df['ZIPCode'].isnull()) &
    (~personal_df['FirstName'].isnull()) &
    (~personal_df['LastName'].isnull())
)

personal_df = personal_df[cond].drop_duplicates()

personal_df
```

# Escritura

```python
personal_df.to_csv('../../1_Datasets/2_Staging/Personal_Data.csv', index=False)
```
