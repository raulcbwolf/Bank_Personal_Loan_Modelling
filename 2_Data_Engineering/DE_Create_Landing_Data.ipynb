# Librerías

```python
import names
import pandas as pd
import random
```

# Lectura

```python
df = pd.read_csv('../1_Datasets/3_Consumption/Bank_Personal_Loan_Modelling.csv')
df.columns
```

# Filtrado de columnas

```python
personal_columns = [
    'ID',
    # 'First Name',
    # 'Last Name',
    'Age',
    'Family',
    'ZIP Code'
]

professional_columns = [
    'ID',
    'Education',
    'Experience',
    'Income'
]

bank_columns = [
    'ID',
    'CCAvg',
    'Mortgage',
    'Personal Loan',
    'Securities Account',
    'CD Account',
    'Online',
    'CreditCard'
]
```

```python
personal_df = df[personal_columns]
professional_df = df[professional_columns]
bank_df = df[bank_columns]
```

# Inclusión de errores en dataset Personal

```python
first_names = [names.get_first_name() for i in range(df.shape[0])]
last_names = [names.get_last_name() for i in range(df.shape[0])]

tmp_df = pd.DataFrame({
    'First Name': first_names,
    'Last Name': last_names
})

personal_df = pd.concat([personal_df, tmp_df], axis=1)
```

```python
personal_df
```

```python
wrong_personal_df = personal_df.sample(2564).copy()
duplicated_df     = personal_df.sample(1234).copy()
```

```python
wrong_personal_df['Age'] = wrong_personal_df.Age.apply(lambda x: random.randint(-50, -1))
wrong_personal_df['Family'] = wrong_personal_df.Age.apply(lambda x: random.randint(-5, -1))

for c in wrong_personal_df.columns:
    wrong_personal_df[c] = wrong_personal_df[c].apply(lambda x: None if random.randint(0, 10) < 3 else x)
```

```python
wrong_personal_df
```

```python
landing_personal_df = pd.concat([personal_df, wrong_personal_df, duplicated_df]).sample(frac = 1)

landing_personal_df
```

```python
landing_personal_df
```

# Solución a los errores del dataset Personal

```python
cond = (
    (landing_personal_df['Age'] >= 0) &
    (~landing_personal_df['Age'].isnull()) &
    (landing_personal_df['Family'] >= 0) &
    (~landing_personal_df['Family'].isnull()) &
    (~landing_personal_df['ZIP Code'].isnull()) &
    (~landing_personal_df['First Name'].isnull()) &
    (~landing_personal_df['Last Name'].isnull())
)

landing_personal_df[cond].drop_duplicates()
```

# Escritura de datos de salida

```python
landing_personal_df.to_csv('../1_Datasets/1_Landing/Personal_Data.csv', index=False)
professional_df.to_csv('../1_Datasets/1_Landing/Professional_Data.csv', index=False)
bank_df.to_csv('../1_Datasets/1_Landing/Bank_Data.csv', index=False)
```
